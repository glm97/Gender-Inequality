```python
import numpy as np
import pandas as pd
```

```python
base = pd.read_csv('adult.csv')
```

```python
base2 = base.drop(['capital.gain', 'capital.loss','native.country'], axis=1)
base2.head()
```

```python
base2.dtypes
```

```python
base2['workclass'] = base2['workclass'].astype('category')
base2['education'] = base2['education'].astype('category')
base2['marital.status'] = base2['marital.status'].astype('category')
base2['occupation'] = base2['occupation'].astype('category')
base2['relationship'] = base2['relationship'].astype('category')
base2['race'] = base2['race'].astype('category')
base2['sex'] = base2['sex'].astype('category')
base2['income'] = base2['income'].astype('category')
base2.dtypes
```

```python
base2['workclass'].cat.categories = ['Doesnt-work', 'Federal-gov', 'Local-gov', 'Never-worked', 'Private',
       'Self-emp-inc', 'Self-emp-not-inc', 'State-gov', 'Without-pay']
base2.head()
```

```python
base2['occupation'].cat.categories = ['Doesnt-work', 'Adm-clerical', 'Armed-Forces', 'Craft-repair', 'Exec-managerial',
       'Farming-fishing', 'Handlers-cleaners', 'Machine-op-inspct',
       'Other-service', 'Priv-house-serv', 'Prof-specialty', 'Protective-serv',
       'Sales', 'Tech-support', 'Transport-moving']
base2.head()
```
